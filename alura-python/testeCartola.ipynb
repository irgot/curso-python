### Populando visitante e posicao

```python
pontos_new = atletas.merge(pontos, on='atleta_id')[['atleta_id','apelido','rodada_id_y','pontos','preco','variacao','media','clube_id','posicao_id']]
pontos_new = pontos_new.merge(right = partidasDF, how='left',  left_on=['rodada_id_y','clube_id'], right_on=['rodada','clube_visitante_id'])[['atleta_id','apelido','rodada_id_y','pontos','preco','variacao','media','clube_id','posicao_id','clube_visitante_id']]
pontos_new['clube_visitante_id'] = pontos_new.clube_visitante_id.fillna(False)
pontos_new['visitante']=pontos_new['clube_visitante_id'].astype(bool)
pontos_new.drop(columns=['clube_visitante_id'],inplace=True)

pd.set_option('display.max_rows', 200)
pontos_new=pontos_new.merge(right = partida_atualDF, how='left',  left_on=['clube_id'], right_on=['clube_visitante_id'])[['atleta_id','apelido','rodada_id_y','pontos','preco','variacao','media','clube_id','posicao_id','visitante','diferenca_posicao']]
pontos_new=pontos_new.merge(right = partida_atualDF, how='left',  left_on=['clube_id'], right_on=['clube_casa_id'])[['atleta_id','apelido','rodada_id_y','pontos','preco','variacao','media','clube_id','posicao_id','visitante','diferenca_posicao_y','diferenca_posicao_x']]
pontos_new['diferenca_posicao_x'].fillna(value=0,inplace=True)
pontos_new['diferenca_posicao_y'].fillna(value=0,inplace=True)
pontos_new['diferenca_posicao']=pontos_new.diferenca_posicao_x+pontos_new.diferenca_posicao_y
pontos_new.drop(columns=['diferenca_posicao_x','diferenca_posicao_y'],inplace=True)
pontos_new = pontos_new.merge(right=atletas, how='left',left_on=['atleta_id'],right_on =['atleta_id'])[['clube_id_y','atleta_id','apelido_y','rodada_id_y','rodada_id','pontos','pontos_num','preco','preco_num','variacao','variacao_num','media','media_num','clube_id_x','posicao_id_x','visitante','status_id','diferenca_posicao']]
pontos_new.head(200)
```
